```python
import ipywidgets as widgets 
```

```python
file = open("cape.jpg", "rb","rb")
image = file.read()
img=widgets.Image(
    value=image,
    format='png',
    width=300,
    height=400,
)
display(img)
```

```python
slide = widgets.IntSlider(
    value=0,
    min=0,
    max=1,
    step=1,
    description='Test:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
display(slide)
```
